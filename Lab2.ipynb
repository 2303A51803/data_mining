```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


```

```python
df = pd.read_csv('/content/lab2.csv', delimiter=';')  # using semicolon as delimiter

# Display the first few rows of the dataset
df.head()
```

```python
df.columns[0]

```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='quality', y='alcohol', data=df, palette='viridis')
plt.title('Bar Plot: Average Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Average Alcohol')
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Scatter plot between alcohol and quality
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df, color='purple')
plt.title('Scatter Plot of Alcohol vs Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```

```python
# Pair plot for a few columns of interest
sns.pairplot(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['alcohol'], bins=20, kde=True, color='blue')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='quality', y='alcohol', data=df, palette='viridis')
plt.title('Violin Plot of Alcohol by Quality')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
quality_counts = df['quality'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=plt.cm.Paired.colors)
plt.title('Distribution of Wine Quality')
plt.show()
```
